### Load Dataset Bersih

```python
import pandas as pd

df = pd.read_csv('../data/Data_bersih2.csv')
df.head()
```

### Korelasi

```python
correlation = df["age"].corr(df["charges"])
print(f"Korelasi antara Umur dengan Rentang adalah {correlation}")

r2 = correlation ** 2
print(f"Koefisien determinasi (R²) antara Umur dan Rentang adalah {r2}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi hubungan antar dua kolom numerik (Scatter Plot)
numerical_cols_scatter = [ "age", "charges"]
sns.scatterplot(data=df, x=numerical_cols_scatter[0], y=numerical_cols_scatter[1], alpha=0.6)
plt.title(f"Scatter Plot of {numerical_cols_scatter[0]} vs {numerical_cols_scatter[1]}")
plt.xlabel(numerical_cols_scatter[0])
plt.ylabel(numerical_cols_scatter[1])
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['age'], df['charges'])
plt.title('Korelasi Antara Umur dan Rentang')
plt.xlabel('Umur')
plt.ylabel('Rentang')
plt.grid(True)
plt.show()
```

### Regression

```python
import statsmodels.formula.api as smf

# Definisikan dan sesuaikan model regresi Age vs Charges
model_experience_salary = smf.ols('charges ~ Q("age")', data=df).fit()

# Mengambil koefisien dari model regresi Pengalaman vs Salary
intercept_experience_salary = model_experience_salary.params['Intercept']
experience_coef_salary = model_experience_salary.params['Q("age")']

# Mengambil jumlah kuadrat Residual (SSR) dari model baru
ssr_experience_salary = model_experience_salary.ssr

print(f"Jumlah Kuadrat Residual (SSR) untuk Rentang vs Umur: {ssr_experience_salary:.4f}")
print(f"Persamaan Garis Regresi (Rentang vs Umur): ")
print(f"Rentang_prediksi = {intercept_experience_salary:.4f} + {experience_coef_salary:.4f} * age")
```
