### Load Dataset

```python
import pandas as pd

df = pd.read_csv('Data_bersih.csv')
df.head()
```

### Korelasi

```python
correlation = df["hours_studied"].corr(df["exam_score"])
print(f"Korelasi antara Lama Belajar dan Hasil Score adalah {correlation}")

r2 = correlation ** 2
print(f"Koefisien determinasi (R²) antara Lama Belajar dan Hasil Score adalah {r2}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi hubungan antar dua kolom numerik (Scatter Plot)
numerical_cols_scatter = [ "hours_studied", "exam_score"]
sns.scatterplot(data=df, x=numerical_cols_scatter[0], y=numerical_cols_scatter[1], alpha=0.6)
plt.title(f"Scatter Plot of {numerical_cols_scatter[0]} vs {numerical_cols_scatter[1]}")
plt.xlabel(numerical_cols_scatter[0])
plt.ylabel(numerical_cols_scatter[1])
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['hours_studied'], df['exam_score'])
plt.title('Korelasi Antara Lama Belajar dan Hasil Skor')
plt.xlabel('Lama Belajar')
plt.ylabel('Hasil Skor')
plt.grid(True)
plt.show()
```

### Regression

```python
import statsmodels.formula.api as smf

# Definisikan dan sesuaikan model regresi Years of Experience vs Salary
model_experience_salary = smf.ols('exam_score ~ Q("hours_studied")', data=df).fit()

# Mengambil koefisien dari model regresi Pengalaman vs Salary
intercept_experience_salary = model_experience_salary.params['Intercept']
experience_coef_salary = model_experience_salary.params['Q("hours_studied")']

# Mengambil jumlah kuadrat Residual (SSR) dari model baru
ssr_experience_salary = model_experience_salary.ssr

print(f"Jumlah Kuadrat Residual (SSR) untuk Hasil Skor vs Lama Belajar: {ssr_experience_salary:.4f}")
print(f"Persamaan Garis Regresi (Hasil Skor vs Lama Belajar): ")
print(f"Pendapatan_prediksi = {intercept_experience_salary:.4f} + {experience_coef_salary:.4f} * hours_studied")
```
