```python
import pandas as pd

df = pd.read_excel("contoh_data_bermasalah.xlsx")
df.head()
```

```python
# Hitung jumlah nilai yang hilang di kolom 'Status' sebelum penghapusan
missing_values_before_drop = df['Status'].isnull().sum()
print(f"Jumlah nilai yang hilang di kolom 'Status' sebelum penghapusan: {missing_values_before_drop}")

# Hapus baris dengan nilai yang hilang di kolom 'Status'
df_dropped = df.dropna(subset=['Status']).copy()

# Verifikasi bahwa tidak ada lagi nilai yang hilang di kolom 'Status'
missing_values_after_drop = df_dropped['Status'].isnull().sum()
print(f"\nJumlah nilai yang hilang di kolom 'Status' setelah penghapusan: {missing_values_after_drop}")

# Tampilkan beberapa baris pertama DataFrame setelah penghapusan
print("\nBeberapa baris pertama DataFrame setelah penghapusan:")
display(df_dropped.head())

print("\nUkuran DataFrame setelah penghapusan nilai yang hilang: {df_dropped.shape}")
```
